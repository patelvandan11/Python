# line plot using seaborn library in python

```python
import seaborn as sns
import matplotlib.pyplot as plt 

var=[1,2,3,4,5,6,7]
var2=[2,3,4,1,5,6,7]

plt.plot(var,var2)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd 

var=[1,2,3,4,5,6,7]
var2=[2,3,4,1,5,6,7]

x=pd.DataFrame({"var":var,"var2":var2})

sns.lineplot(x="var",y="var2",data=x)

plt.show()
```
