# pair plot 

```python
import matplotlib.pyplot as plt 
import pandas as pd 
import seaborn as sns 
```

```python
var=pd.read_csv("tips.csv")
var
```

```python
sns.pairplot(var)
plt.show()
```

```python
sns.pairplot(var,hue="sex")
plt.show()
```

```python
sns.pairplot(var,vars=["total_bill","tip"],hue="sex")
plt.show()
```

```python
sns.pairplot(var,vars=["total_bill","tip"],hue="sex",hue_order=["Male","Female"])
plt.show()
```

```python
sns.pairplot(var,hue="sex",hue_order=["Male","Female"],y_vars=["total_bill","tip"])
plt.show()
```

```python
sns.pairplot(var,hue="sex",hue_order=["Male","Female"],x_vars=["total_bill","tip"])
plt.show()
```

```python
sns.pairplot(var,hue="sex",hue_order=["Male","Female"],x_vars=["total_bill","tip"],kind="kde")   # "scatter"    "kde"   "hist"  "reg"
plt.show()
```

```python
sns.pairplot(var,hue="sex",hue_order=["Male","Female"],y_vars=["total_bill","tip"],kind="kde",diag_kind="hist")   # "scatter"    "kde"   "hist"  "reg"
plt.show()
```

```python
sns.pairplot(var,hue="sex",hue_order=["Male","Female"],markers=["*","o"])   # "scatter"    "kde"   "hist"  "reg"
plt.show()
```
