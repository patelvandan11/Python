# data changes :: color & font

```python
var=np.linspace(1,10,20).reshape(4,5)
var
```

```python
y={"fontsize":10,"color":"r"}
sns.heatmap(var,vmin=0,vmax=10,annot=True,cmap="icefire",annot_kws=y)
```

```python
var=np.linspace(1,10,20).reshape(4,5)
var
```

```python
y={"fontsize":10,"color":"r"}
# sns.heatmap(var,vmin=0,vmax=10,annot=True,cmap="icefire",annot_kws=y,linewidths=10)  
# sns.heatmap(var,vmin=0,vmax=10,annot=True,cmap="icefire",annot_kws=y,linewidths=10,linecolor="m")
# sns.heatmap(var,vmin=0,vmax=10,annot=True,cmap="icefire",annot_kws=y,linewidths=10,linecolor="m",cbar=False)
# sns.heatmap(var,vmin=0,vmax=10,annot=True,cmap="icefire",annot_kws=y,linewidths=10,linecolor="m",cbar=False,xticklabels=False)
# sns.heatmap(var,vmin=0,vmax=10,annot=True,cmap="icefire",annot_kws=y,linewidths=10,linecolor="m",cbar=False,yticklabels=False)
sns.heatmap(var,vmin=0,vmax=10,annot=True,cmap="icefire",annot_kws=y,linewidths=10,linecolor="m",cbar=False,yticklabels=False,xticklabels=False)
plt.show()
```

```python
b=sns.heatmap(var,vmin=0,vmax=10,annot=True,cmap="icefire",annot_kws=y,linewidths=10,linecolor="m",cbar=False,yticklabels=False,xticklabels=False)
b.set(xlabel="v'",ylabel="v")
sns.set(font_scale=20)
plt.show()
```
