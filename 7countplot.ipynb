# count  plot

* difference between countplot and barplot 

    * countplot plots the count of th number of recordes by category .
    * barplot plots a value of metric for each category (by default , barplot plots the mean of a variable ,by category.).

```python
import matplotlib.pyplot as plt 
import pandas as pd 
import seaborn as sns
```

```python
var=pd.read_csv("tips.csv")
var
```

```python
sns.countplot(x="sex",data=var)                                                                                
plt.show()                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                       
```

```python
sns.barplot(x="sex",y="size",data=var)
plt.show()
```

```python
sns.countplot(x="sex",data=var, hue="smoker")
plt.show()
```

```python
sns.countplot(y="sex",data=var, hue="smoker")
plt.show()
```

```python
sns.countplot(x="sex",data=var, hue="smoker",palette="icefire")
plt.show()
```

```python
sns.countplot(x="sex",data=var, hue="smoker",color="m")
sns.countplot(x="sex",data=var, hue="smoker",color="m",saturation=0.9)



plt.show()
```
