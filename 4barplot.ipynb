# barplot

```python
import matplotlib.pyplot as plt 
import pandas as pd 
import seaborn as sns 
deta=pd.read_csv("penguins.csv")
deta.head(20) 
```

```python
sns.barplot(x=deta.island,y=deta.bill_length_mm,data=deta)
```

```python
sns.barplot(x=deta.island,y=deta.bill_length_mm,data=deta)
plt.show()    #  delete ::<Axes: xlabel='island', ylabel='bill_length_mm'>
```

```python
sns.barplot(x="island",y="bill_length_mm",data=deta)
```

```python
sns.barplot(x="island",y="bill_length_mm",data=deta,hue="sex")  # saprate a male and female parameters 
```

```python
sns.barplot(x="island",y="bill_length_mm",data=deta,hue="sex",order=["Dream","Torgersen","Biscoe"])  # order= :" to change the order of bar plot"
```

```python
sns.barplot(x="island",y="bill_length_mm",data=deta,hue="sex",order=["Dream","Torgersen","Biscoe"],hue_order=["FEMALE","MALE"]) 
# hue_order=  :"to change the order of a hue parameter"
```

```python
 sns.barplot(x="island",y="bill_length_mm",data=deta,hue="sex",order=["Dream","Torgersen","Biscoe"],hue_order=["FEMALE","MALE"],ci=1) 
 #ci  black line size 
```

```python
sns.barplot(x="bill_depth_mm",y="bill_length_mm",data=deta,orient="h")
plt.show()
```

```python
 sns.barplot(x="island",y="bill_length_mm",data=deta,hue="sex",order=["Dream","Torgersen","Biscoe"],hue_order=["FEMALE","MALE"],color="b") 
```

```python
 sns.barplot(x="island",y="bill_length_mm",data=deta,color="b") 
```

```python
 sns.barplot(x="island",y="bill_length_mm",data=deta,hue="sex",color="b",palette="icefire") 
```

```python
 sns.barplot(x="island",y="bill_length_mm",data=deta,hue="sex",saturation=0.1) 
```

```python
 sns.barplot(x="island",y="bill_length_mm",data=deta,hue="sex",color="b",palette="icefire",errcolor="m") 
```

```python
 sns.barplot(x="island",y="bill_length_mm",data=deta,hue="sex",color="b",palette="icefire",errcolor="m",capsize=0.2) 
```

```python
 sns.barplot(x="island",y="bill_length_mm",data=deta,hue="sex",color="b",palette="icefire",errcolor="m",dodge=False) 
```

```python
 sns.barplot(x="island",y="bill_length_mm",data=deta,hue="sex",color="b",palette="icefire",errcolor="m",alpha=0.5 ) 
 #  alpha= use a graph of lite color  


```

```python
sns.set(style="darkgrid")
sns.barplot(x="island",y="bill_length_mm",data=deta,hue="sex",color="b",palette="icefire",errcolor="m" ) 
 #  alpha= use a graph of lite color  
```
