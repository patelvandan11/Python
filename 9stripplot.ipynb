# strip plot

* strip plot is very simple to understnd. ti is basically a scatter plot that differentiates different categories .
    strip plots are considered a good alternative to a box plot or a violin plot.


```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
var=pd.read_csv("tips.csv")
var
```

```python

# sns.stripplot(x="day",y="total_bill",data=var,hue="sex")
# sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="rocket_r",linewidth=1)
# sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="rocket_r",linewidth=1,edgecolor="g")
# sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="rocket_r",linewidth=1,edgecolor="g",jitter=True)
# sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="rocket_r",linewidth=1,edgecolor="g",jitter=True,size=6)
# sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="rocket_r",linewidth=1,edgecolor="g",jitter=True,size=6,marker="<")
# sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="rocket_r",linewidth=1,edgecolor="g",jitter=True,size=6,marker="<",alpha=0.5)
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="rocket_r",linewidth=1,edgecolor="g",jitter=True,size=6,marker="<",alpha=0.5,)
plt.show()
```
