# scatter plot

```python
import matplotlib.pyplot as plt 
import pandas as pd 
import seaborn as sns 

d=pd.read_csv("penguins.csv")
d=d.head(20)
d
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=d)
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=d,hue="sex")
plt.show()
```
