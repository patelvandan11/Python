# box plot 

```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
var=pd.read_csv("tips.csv")
var
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var)
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var,hue="time")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var,hue="sex")
plt.show()
```

```python

sns.boxplot(x="day",y="total_bill",data=var,hue="sex",order=["Fri","Sun","Thur","Sat"],showmeans=True,meanprops={"marker":"*","markeredgecolor":"m"},linewidth=3,palette="icefire",orient="v")
```
